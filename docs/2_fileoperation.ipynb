# 基本文件操作
这一章以项目文件夹下的“zhiyaopingfan.dv”（只要平凡，作曲：黄超，作词：格格）为例，介绍dvfile的基本文件操作。
## 打开文件
首先导入dvfile库

```python
import dvfile
```

用opendv打开文件，返回一个dv文件对象

```python
d=dvfile.opendv("zhiyaopingfan.dv")
type(d)
```

可以用print将dv文件对象打印出来以便快速查看。

注意：dvfile库中各种对象，print输出的内容在后续版本的dvfile中随时可能调整与优化，仅用于命令行交互与程序调试，不建议解析字符串后继续利用数据。

```python
print(d)
```

## 保存文件
dv文件对象可以用save方法保存为dv文件

```python
d.save("zhiyaopingfan2.dv")
```

## 文件格式转换
dv文件对象可以导出ust、nn、mid、musicxml文件，需要安装好对应文件的python库

```python
d.to_ust_file()[0].save("zhiyaopingfan.ust")#按音轨导出ust文件对象列表（需要utaufile）
d.to_nn_file()[0].save("zhiyaopingfan.nn")#按音轨导出nn文件对象列表（需要utaufile）
d.to_midi_file().save("zhiyaopingfan.mid")#导出mid（需要mido）
d.to_music21_score().write("xml",fp="zhiyaopingfan.xml")#导出musicxml（需要utaufile和music21）
```

dvfile在导出文件时，如果目标文件不支持汉字拼音双重歌词（ust、music21），则默认使用拼音歌词以保证对大多数歌声合成软件的兼容性。如果需要导出汉字歌词，请使用use_hanzi=True参数

注意：mido库只支持ascii字符，因此无法导出汉字mid文件

```python
d.to_ust_file(use_hanzi=True)[0].save("zhiyaopingfan.mid")
```

## 导出五线谱

```python
d.to_music21_score(use_hanzi=True).show()
```

由于music21库很大，第一次运行这行代码需要等待几秒钟，然后会打开musescore窗口并显示五线谱，可以导出为图片或pdf文件。
